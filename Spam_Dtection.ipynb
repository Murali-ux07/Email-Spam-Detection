```python
import pandas as pd
```

```python
import re
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
```

```python
nltk.download("stopwords")
```

```python
stemmer = PorterStemmer()
```

```python
stop_words = set(stopwords.words("english"))
```

```python
df = pd.read_csv("/content/drive/MyDrive/Spam_Data.csv", encoding = 'latin-1')
```

```python
df.head(10)
```

```python
def preprocessed_text(text):
    # Replace non-word characters with space, removing extra spaces
    text = re.sub(r'\W+', ' ', text)
    # Convert to lowercase
    text = text.lower()
    # Tokenize the text into words
    words = text.split()
    # Remove stop words, stem words, and ensure meaningful tokens only
    words = [stemmer.stem(word) for word in words if len(word) > 1 and word not in stop_words]
    # Join back into a single string
    return ' '.join(words)
```

```python
df["cleaned_message"] = df["Message"].apply(preprocessed_text)
```

```python
df.head()
```

```python
df.tail(12)
```

```python
df.columns = ['label', 'Message'] # Assign the correct number of column names (2) to the existing columns of the DataFrame.
df['cleaned_message'] = '' # Creating a new column named 'cleaned_message' and give temporary value.
df["cleaned_message"] = df["Message"].apply(preprocessed_text)  # Apply the preprocessing function as before.
df.columns = ['label', 'Message', 'cleaned_message'] # Now, you can rename the column.
```

```python
df['label'] = df["label"].map({'ham': 0, 'spam': 1})
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
vectorizer = TfidfVectorizer(max_features=3000)
```

```python
df.head()
```

```python
x = vectorizer.fit_transform(df["cleaned_message"])
```

```python
y = df["label"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")
```

```python
def predict_email(email_text):
  processed_data = preprocessed_text(email_text)
  vectorized_text = vectorizer.transform([processed_data]) # Pass processed_data instead of preprocessed_text function
  prediction = model.predict(vectorized_text)
  return "Spam" if prediction[0] == 1 else "No Spam"
```
